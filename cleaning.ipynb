```python
import pandas as pd
import numpy as np
import os
```

```python
data = pd.read_csv('./amazon_scrape_results/amazon_india_soft_toys_detailed.csv')
```

```python
data.head(5)
```

```python
df=data
```

```python
df = df.drop_duplicates(subset=["Product URL"])
```

```python
df["Title"] = df["Title"].fillna("N/A")
df["Brand"] = df["Brand"].fillna("N/A")
df["Product URL"] = df["Product URL"].fillna("N/A")
df["Image URL"] = df["Image URL"].fillna("N/A")
```

```python
df.head(5)
```

```python
df["Selling Price"] = df["Selling Price"].astype(str).str.replace("₹", "", regex=False).str.replace(",", "").str.strip()
df["Selling Price"] = pd.to_numeric(df["Selling Price"], errors="coerce")
```

```python
df.head(5)
```

```python
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")
df["Reviews"] = pd.to_numeric(df["Reviews"], errors="coerce")
```

```python
df["Selling Price"] = df["Selling Price"].fillna(0)
df["Rating"] = df["Rating"].fillna(0)
df["Reviews"] = df["Reviews"].fillna(0)
```

```python
df.head(5)
```

```python
# Remove duplicates based on the "Product URL" column
df = df.drop_duplicates(subset=["Product URL"])
```

```python
len(df)
```

```python
output_file_path = os.path.join('amazon_india_soft_toys_cleaned.csv')
df.to_csv(output_file_path, index=False)
print(f"Data cleaned and saved to {output_file_path}")
```
